```python
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
sns.kdeplot(penguins.bill_length_mm);
```

```python
sns.kdeplot(penguins.bill_length_mm, shade=True);
```

```python
sns.kdeplot(penguins.bill_length_mm, penguins.bill_depth_mm);
```

```python
sns.kdeplot(penguins.bill_length_mm, penguins.bill_depth_mm, 
           n_levels=20);
```

```python
sns.kdeplot(penguins.bill_length_mm, penguins.bill_depth_mm, 
           n_levels=20, shade=True);
```

```python
sns.kdeplot(penguins.bill_length_mm, penguins.bill_depth_mm, 
           n_levels=20, shade=True, cbar=True);
```
