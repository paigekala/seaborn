```python
import numpy as np
import pandas as pd
import seaborn as sns

from matplotlib import pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
penguins.head()
```

```python
sns.distplot(penguins.bill_length_mm);
```

```python
plt.rcParams["patch.force_edgecolor"] = True
```

```python
sns.distplot(penguins.bill_length_mm, kde=False);
```

```python
sns.distplot(penguins.bill_length_mm, bins=5)
```

```python
sns.distplot(penguins.bill_length_mm, bins=10)
```

```python
from scipy.stats import norm
sns.distplot (penguins.bill_length_mm, fit=norm);
```

```python
from scipy.stats import skewnorm
sns.distplot (penguins.bill_length_mm, fit=skewnorm, kde=False);
```

```python
sns.distplot(penguins.bill_length_mm, kde=False, hist_kws={'alpha' :1});
```

```python
sns.distplot(penguins.bill_length_mm, kde=False,
            hist_kws={'alpha':1, 'histtype': 'step'});
```

```python
sns.distplot(penguins.bill_length_mm, kde_kws={'lw':3});
```

```python
sns.distplot(penguins.bill_length_mm, kde_kws={'lw':3},
             rug=True, rug_kws={'height':.10});
```
